```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#converting categorical data into numerical data
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent=dataset[["charges"]]
```

```python
independent
```

```python
dependent
```

```python
# Now We can split the training set and the test set
# We can use the same code as in the "SIMPLE LINEAR REGRESSION"
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent,dependent, test_size = 0.30,random_state = 0)

#NOTE:
# After the execution of r2_score the result was -0. THis is not near to 1. So we have to do the standardization method form here itself
```

```python
X_train
```

```python
#Standardization Technique: THis technique will be used only after the the result is not nearly to 1.
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test) #After this execution with this formula "X-Mue/sigma" again it'll save the calculation in X_train
```

```python
X_train
```

```python
!pip install xgboost
```

```python
#Model Creation
from xgboost import XGBRegressor

# Create the model
regr = XGBRegressor(
    n_estimators=100,   # Number of boosting rounds
    random_state=0,     # For reproducibility
    learning_rate=0.1,  # Step size shrinkage
    max_depth=3,        # Depth of each tree
    objective='reg:squarederror') # Loss function for regressio
regr.fit(X_train,y_train) #Training sets input and output, Hre only the MODEL gonna create.
```

```python
#Evaluation Metrics
y_pred=regr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
# Here we got -0, Not nearly to 1. So our predicted model is not a good model.
#So we have to go with the standardization method to predict the exact result.
```

```python
#We can use Pickle Library - Name it as "finalized model.
import pickle
filename = 'finalized_model_XGBoost.sav'
pickle.dump(regr, open(filename,'wb'))
```
